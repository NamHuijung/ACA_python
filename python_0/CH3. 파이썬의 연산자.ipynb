```python
# 02. 기본 연산자
'''
파일명 : Ex04-1.py
개요 : 두 변수에 저장된 값의 산술 연산 결과를 출력
작성자 : 남희정
최초작성일 : 2021-08-18
'''

a = 7
b = 2
print('{} + {} = {}'.format(a, b, a + b))
print('{} - {} = {}'.format(a, b, a - b))
print('{} * {} = {}'.format(a, b, a * b))
print('{} ** {} = {}'.format(a, b, a ** b))
print('{} / {} = {}'.format(a, b, a / b))
print('{} // {} = {}'.format(a, b, a // b))
print('{} % {} = {}'.format(a, b, a % b))
```

```python
# 02. 기본 연산자
'''
파일명 : Ex04-2.py
개요 : 변수 a와 b에 10과 20을 저장한 뒤 저장된 값을 교환하는 프로그램
작성자 : 남희정
최초작성일 : 2021-08-18
'''

a, b = 10, 20
print('a = %d, b = %d' %(a, b))
a, b = b, a
print('a = %d, b = %d' %(a, b))
```

```python
# 02. 기본 연산자
'''
파일명 : Ex04-3.py
개요 : 저금통에 용돈을 넣고 빼는 과정을 보여 주는 프로그램
작성자 : 남희정
최초작성일 : 2021-08-18
'''

piggy_bank = 0

money = 10000
piggy_bank += money
print('저금통에 용돈 {}원을 넣었습니다.'.format(money))
print('지금 저금통에는 {}원이 있습니다.'.format(piggy_bank))

snack = 2000
piggy_bank -= snack
print('저금통에서 스낵 구입비 {}원을 뺐습니다.'.format(snack))
print('지금 저금통에는 {}원이 있습니다.'.format(piggy_bank))
```

```python
# 02. 기본 연산자
'''
파일명 : Ex04-4.py
개요 : 관계 연산자의 결과를 출력
작성자 : 남희정
최초작성일 : 2021-08-18
'''

a = 15
print('{} > 10 : {}'.format(a, a > 10))
print('{} < 10 : {}'.format(a, a < 10))
print('{} >= 10 : {}'.format(a, a >= 10))
print('{} <= 10 : {}'.format(a, a <= 10))
print('{} == 10 : {}'.format(a, a == 10))
print('{} != 10 : {}'.format(a, a != 10))
```

```python
# 02. 기본 연산자
'''
파일명 : Ex04-5.py
개요 :  논리 연산자의 결과를 출력
작성자 : 남희정
최초작성일 : 2021-08-18
'''

a = 10
b = 0
print('{} > 0 and {} > 0 : {}'.format(a, b, a>0 and b>0))
print('{} > 0 or {} > 0 : {}'.format(a ,b, a>0 or b>0))
print('not {} : {}'.format(a, not a))
print('not {} : {}'.format(b, not b))
```

```python
# 02. 기본 연산자
'''
파일명 : Ex04-6.py
개요 :  비트 연산자의 결과를 출력
작성자 : 남희정
최초작성일 : 2021-08-18
'''

a = 10 # 10 = 0000 0010
b = 70 # 70 = 0100 0110
print('a & b : {}'.format(a & b))
print('a | b : {}'.format(a | b))
print('a ^ b : {}'.format(a ^ b))
print('~a : {}'.format(~a))
print('a << 1 : {}'.format(a << 1))
print('a >> 1 : {}'.format(a >> 1))
```

```python
# 02. 기본 연산자
'''
파일명 : Ex04-7.py
개요 :  '#'과 ''를 이용하여 크리스마스 트리 형태를 출력
작성자 : 남희정
최초작성일 : 2021-08-18
'''

tree = '#'
space = ' '
print(space * 4 + tree * 1)
print(space * 3 + tree * 3)
print(space * 2 + tree * 5)
print(space * 1 + tree * 7)
print(space * 0 + tree * 9)
```

```python
# 02. 기본 연산자
'''
파일명 : Ex04-8.py
개요 :  조건 연산자를 이용하여 두 수의 차이를 절댓값으로 출력
작성자 : 남희정
최초작성일 : 2021-08-18
'''

a = 10
b = 20
result = (a-b) if (a >= b) else (b-a)
print('{}과 {}의 차이는 {}입니다.'.format(a, b, result))
```

```python
# 1. 임의의 두 자리 정수(10~99)를 입력받아서 십의 자리와 일의 자리로 분리하여 출력하는 프로그램

num = int(input('10 ~ 99 사이의 정수를 입력하세요 >> '))
print('십의 자리 : {}'.format(num // 10))
print('일의 자리 : {}'.format(num % 10))
```

```python
# 2. 1분은 60초이고, 1시간은 60분입니다. 따라서 1시간은 3600초입니다.
# 임의의 초를 입력받아 해당 초를 시, 분, 초로 변환하여 출력하는 프로그램

time = int(input('초를 입력하세요 >> '))
hour = time // 3600
min = (time % 3600) // 60
sec = (time % 3600) % 60
print('변환 결과는 {0}시간 {1}분 {2}초입니다.'.format(hour, min, sec))
```

```python
# 3. 네 자리 정수로 구성된 사원번호를 기준으로 근무 시간을 결정하고자 합니다.
# 사원번호의 끝자리 숫자가 5이상이면 '오전', 아니면 '오후'를 출력하는 프로그램

num = int(input('4자리 사원번호를 입력하세요 >> '))
res = "오전" if (num % 10) >= 5 else "오후"
print('근무 시간은 {}입니다.'.format(res))
```

```python
# 4. 1박스에 라면이 20개씩 들어간다. 라면이 21개가 있다면 2박스가 있어야 모두 담을 수 있다.
# 라면의 개수를 입력받아 라면을 담기 위해서 필요한 박스의 개수를 구하여 출력하는 프로그램

print('한 박스에 20개의 라면을 담을 수 있습니다.')
print('라면의 개수를 입력하시면 필요한 박스 수를 알려드립니다.')

count = int(input('라면의 개수를 입력하세요 >> '))

res = (count // 20) if (count % 20) == 0 else ((count // 20) +1)

print('{0}개 라면을 담으려면 {1}박스가 필요합니다.'.format(count, res))
```

```python
# 5. 국어, 영어, 수학 점수를 각각 입력받아서 평균을 구하고,
# 평균이 80점 이상이면 '합격', 아니면 '불합격'을 출력하는 프로그램

kor = int(input('국어 점수를 입력하세요 >> '))
eng = int(input('영어 점수를 입력하세요 >> '))
math = int(input('수학 점수를 입력하세요 >> '))

mean = (kor + eng + math) / 3

res = "합격" if mean >= 80 else "불합격"

print('평균은 {0}점이고, 결과는 {1}입니다.'.format(mean, res))
```
