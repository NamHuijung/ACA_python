```python
# 02. while 문
'''
파일명 : Ex06-1.py
개요 : 1부터 10 사이의 모든 정수를 역순으로 출력하는 프로그램
작성자 : 남희정
최초작성일 : 2021-08-19
'''

n = 10 # 초깃값
while n >= 1 : # 종료값
  print(n)
  n -= 1 # 증감값
```

```python
# 02. while 문
'''
파일명 : Ex06-2.py
개요 : 사용자로부터 임의의 정수를 입력받아 모두 리스트에 보관합니다.
      단, 사용자가 0을 입력하면 더 이상 입력받지 않고 프로그램을 종료합니다. 이때 0은 리스트에 보관하지 않는다.
작성자 : 남희정
최초작성일 : 2021-08-19
'''

my_list = []

n = int(input('정수를 입력하세요(종료는 0입니다.) >> '))

while n != 0 :
  my_list.append(n)
  n = int(input('정수를 입력하세요(종료는 0입니다.) >> '))

print(my_list)
```

```python
# 02. while 문
'''
파일명 : Ex06-3.py
개요 : 구구단 2단부터 9단까지 출력하는 프로그램
작성자 : 남희정
최초작성일 : 2021-08-19
'''

dan = 2
while dan <= 9 :
  n = 1
  while n <= 9 :
    print('{} X {} = {}'.format(dan, n, dan * n))
    n += 1
  dan += 1
```

# section07. 반복문 for

```python
# 02. 시퀀스와 for문
'''
파일명 : Ex07-1.py
개요 : 문자열(str) 자료형의 비밀번호를 입력받아서 숫자와 문자가 모두 포함되었는지 확인한 뒤
      모두 포함되어 있으면 '가능한 비밀번호입니다.'를 출력하고, 아니면 '불가능한 비밀번호입니다.'를 출력하는 프로그램
작성자 : 남희정
최초작성일 : 2021-08-19
'''

pwd = input('비밀번호를 입력하세요 >> ')

ch_count = 0
num_count = 0

for ch in pwd :
  if ch.isalpha() : # ch.isalpha() = ch가 문자인 경우 True
    ch_count += 1
  elif ch.isnumeric() : # ch.numeric() = ch가 숫자인 경우 True
    num_count += 1

if (ch_count > 0) and (num_count > 0) :
  print('가능한 비밀번호입니다.')
else : 
  print('불가능한 비밀번호입니다.')
```

```python
# 02. 시퀀스와 for문
'''
파일명 : Ex07-2.py
개요 : 출력하고자 하는 구구단을 입력받아 해당 구구단을 출력하는 프로그램
작성자 : 남희정
최초작성일 : 2021-08-20
'''

dan = int(input('출력할 구구단을 입력하세요 >> '))
for n in range(1, 10) :
  print('{} X {} = {}'.format(dan, n, dan * n))
```

```python
# 5. 1부터 99 사이의 모든 정수를 대상으로 369 게임의 결과를 출력

list = []
for i in range(1, 100) :
  if (i < 10) or (i % 10 == 0) :
    if i % 3 == 0 :
      list.append('짝')
    else : 
      list.append(i)

  else :
    if (i // 10) % 3 == 0 and (i % 10) % 3 == 0:
      list.append('짝짝')
    elif (i // 10) % 3 == 0 or (i % 10) % 3 == 0 :
      list.append('짝')
    else :
      list.append(i)

for i in range(1, 100) :
  if i % 10 != 0 :
    print('{}'.format(list[i-1]), end = '\t')
  else :
    print('{}\t'.format(list[i-1]))
```

# section08. 기타 제어문

```python
# 01. break문
'''
파일명 : Ex08-1.py
개요 : 대한민국의 수도를 맞히는 퀴즈를 구현한 프로그램
      정답을 맞힐 때까지 프로그램은 종료되지 않는다.
작성자 : 남희정
최초작성일 : 2021-08-21
'''

while True :
  city = input('대한민국의 수도는 어디인가요? >> ')
  if city == '서울' or city == 'SEOUL' or city == 'seoul' :
    print('정답입니다.')
    break
  else :
    print('오답입니다. 다시 시도하세요.')
```

```python
# 01. break문
'''
파일명 : Ex08-2.py
개요 : 사용자들의 취미를 입력받아서 리스트에 저장하는 프로그램
      더 이상 취미가 없으면 사용자들은 취미를 입력하지 않고 엔터를 눌러 프로그램을 종료
작성자 : 남희정
최초작성일 : 2021-08-21
'''

hobbies = []
while True :
  hobby = input('취미를 입력하세요(종료는 그냥 엔터) >> ')
  if hobby == '':
    print('입력된 취미가 모두 저장되었습니다.')
    break
  else :
    hobbies.append(hobby)

print(hobbies)
```

```python
# 02. continue문
'''
파일명 : Ex08-3.py
개요 : 사용자로부터 과일 이름을 입력받아 리스트에 보관하는 프로그램
      총 5개의 과일 이름을 보관하는데 리스트에는 '사과', '감귤'을 미리 입력해 두었다.
      중복된 과일은 저장할 수 없으며 만약 중복된 과일 이름을 입력하면 다른 과일을 입력받는다.
작성자 : 남희정
최초작성일 : 2021-08-21
'''

fruits = ['사과', '감귤']
count = 3

while count > 0 :
  fruit = input('어떤 과일을 저장할까요? >> ')
  if fruit in fruits :
    print('동일한 과일이 있습니다.')
    continue
  fruits.append(fruit)
  count -= 1
  print('입력이 {}번 남았습니다.'.format(count))

print('5개의 과일은 {}입니다.'.format(fruits))
```

```python
# 02. continue문
'''
파일명 : Ex08-4.py
개요 : 사용자로부터 5개의 양의 정수를 입력받아 그 합계를 구하는 프로그램
      만약 0 이하의 값이 입력되면 사용자에게 재입력을 요구한다.
작성자 : 남희정
최초작성일 : 2021-08-21
'''

count = 0
total = 0
while count < 5 :
  n = int(input('{}번째 정수를 입력하세요 >> '.format(count + 1)))
  if n <= 0 :
    print('0 이하의 정수는 처리할 수 없습니다.')
    continue
  total += n
  count += 1
print('입력된 5개의 양수의 총 합은 {}입니다.'.format(total))
```

```python
# 1. 현재 10000원을 가지고 있습니다. 
# 얼마를 사용할 것인지 반복해서 입력받아 10000원을 모두 사용하세요.
# 0 이하의 금액은 사용할 수 없으며, 현재 가지고 있는 돈보다 더 큰 금액도 사용할 수 없습니다.

money = 10000
print('현재 {}원이 있습니다.'.format(money))

while money != 0 :
  use = int(input('사용할 금액 입력 >> '))
  if money < use :
    print('{}원이 부족합니다.'.format(use - money))
    continue
  elif use < 0 :
    print('0 이하의 금액은 사용할 수 없습니다.')
    continue
  money -= use
  print('현재 {}원이 있습니다.'.format(money))
```

```python
# 2. 영화 평점을 1부터 5 사이의 정수로 입력받아서 해당 평점만큼 ★을 표시하는 프로그램
# 표시할 수 있는 평점의 범위를 벗어나면 재입력을 요구하세요.

while True :
  user = int(input('이번 영화의 평점을 입력하세요 >> '))
  if user < 1 or user > 5 :
    print('평점은 1 ~ 5 사이만 입력할 수 있습니다.')
  else :
    print('평점 : {}'.format('★'*user))
    break
```

```python
# 3. 저장된 비밀번호를 맞히는 프로그램
# 저장된 비밀번호는 'qwerty'이며 최대 5번 시도할 수 있다.
# 5번 이내에 맞히면 '비밀번호를 맞혔습니다.'를 출력하고 그렇지 않으면 '비밀번호 입력 횟수를 초과했습니다.'를 출력

count = 0
pw = 'qwerty'
while count < 5 :
  user = input('비밀번호를 입력하세요 >> ')
  if pw == user :
    print('비밀번호를 맞혔습니다.')
    break
  elif count == 4 :
    print('비밀번호 입력 횟수를 초과했습니다.')
  count += 1
```

```python
# 4. 다음 조건을 만족하는 구구단을 출력
# 짝수인 단(2, 4, 6,8단)은 출력하지 말고, blank line만 추가
# 각 단까지만 출력하세요.
# 예를 들어 3단은 3X3까지만 출력하고, 5단은 5X5까지만 출력

dan = 2
while dan < 10 :
  if dan % 2 == 0 :
    print()
  else :
    for i in range(1, 10) :
      if i != (dan + 1) :
        print('{} X {} = {}'.format(dan, i, dan * i))
      else : 
        break
  dan += 1
```
