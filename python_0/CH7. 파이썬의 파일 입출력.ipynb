# section 13. 파일 입출력의 이해

- 파일 열기
      파일객체 = open(파일명, 모드)
- 모드
 - r : 읽기, 파일이 없을 때 오류 발생
 - w : 쓰기, 새로 생성
 - a : 추가, 파일이 없을 때는 새로 생성
 - x : 배타적 추가,  파일이 없을 때 새로 생성, 있으면 오류 발생
- 파일 닫기
      파일객체.close()

**모드**
- rt : 텍스트 파일 읽기 모드
- rb : 바이너리 파일 읽기 모드
- wt : 텍스트 파일 쓰기 모드
- wb : 바이너리 파일 쓰기 모드
- at : 텍스트 파일 추가 모드
- ab : 바이너리 파일 추가 모드
- xt : 텍스트 파일 배타적 추가 모드
- xb : 바이너리 파일 배타적 추가 모드

```python
# 02. 파일 출력(output)
'''
파일명 : Ex13-1.py
개요 : 오늘의 스케줄을 입력하면 그 내용을 모두 파일에 보관하는 프로그램
      스케줄을 입력하지 않고 'enter'를 누르면 프로그램 종료
      생성되는 파일의 이름은 현재 날짜이고 확장자는 txt
      '2020-10-22.txt'와 같은 형식을 갖추고 있음
작성자 : 남희정
최초작성일 : 2021-09-07
'''

import time # time 모듈 import

file = open(time.strftime('%Y-%m-%d') + '.txt', 'at') # strftime() : 날짜에 형식을 지정한 문자열을 반환
while True :
  schedule = input('오늘의 스케줄을 입력하세요 >> ')
  if not schedule : # if schedule == '' :
    break
  file.write(schedule + '\n')
file.close()
```

**텍스트 파일 읽기**
1. read() : 파일로부터 데이터를 읽어 들이는 메소드
2. readline() : 텍스트 파일을 한 줄씩 읽어서 처리하는 메소드  
만약 파일이 종료되어 더 이상 읽어 들일 데이터가 없으면 빈 문자열(' ')을 읽어 들인다
3. readlines() : 라인(line) 하나가 아니라 전체 라인(lines)을 모두 읽어 각 라인(line) 단위로 리스트에 저장하는 메소드

```python
# 03. 파일 입력(input)
'''
파일명 : Ex13-2.py
개요 : 동요 '엄마돼지 아기돼지'의 가사가 저장되어 있는 '엄마돼지아기돼지.txt' 파일이 웹하드에 업로드되어 있다.
      이 파일에서 '꿀'이란 글자가 몇 번 나오는지 찾는 프로그램
작성자 : 남희정
최초작성일 : 2021-09-07
'''

file = open('엄마돼지아기돼지.txt', 'rt')

line_list = file.readlines()

count = 0
for line in line_list :
  for ch in line :
    if ch == '꿀' :
      count += 1

print('꿀은 전체 {}번 나타납니다.'.format(count))
```
