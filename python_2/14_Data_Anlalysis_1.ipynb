```python
# 연도와 월은 연산에 사용되는 데이터가 아니라 구분에 사용되는 데이터이므로 문자열 형태로 변환한다.
pre_sale['연도'] = pre_sale['연도'].astype(str)
pre_sale['월'] = pre_sale['월'].astype(str)
```

```python
# 분양가격 데이터 타입을 숫자로 변경하고 평당 분양가격을 계산한다.
# 단위 문자는 'ㄹ+한자키' 사용
# pre_sale['분양가격(㎡)'] = pre_sale['분양가격(㎡)'].astype(float) # 오류

#to_numeric() 함수의 errors 속성을 coerce로 지정하면 오류가 발생된 데이터를 누락값으로 변경한다. 
pre_sale['분양가격(㎡)'] = pd.to_numeric(pre_sale['분양가격(㎡)'], errors='coerce')

pre_sale['평당분양가격'] = pre_sale['분양가격(㎡)'] * 3.3
pre_sale.dtypes
```

```python
# pd.options.display.float_format을 사용해서 데이터프레임에 저장된 실수 데이터의 출력 서식을 지정할 수 있다.
# 출력 서식 '{:,.1f}'는 천 단위마다 ','를 출력하고 '.'뒤의 숫자만큼 소수점 아래로 표현한다.
# 'f'를 붙여주지 않으면 실수가 'e'를 사용하는 지수 형태로 표현된다.
pd.options.display.float_format = '{:,.1f}'.format
```

```python
# groupby() 함수와 피벗 테이블(T) 활용하기

# groupby() 함수로 데이터프레임에 저장된 숫자 데이터의 요약 통계량을 확인할 수 있다.
pre_sale.groupby(pre_sale['연도']).describe()
```

```python
# describe() 함수의 실행 결과에 'T' 속성을 지정하면 요약 통계량이 전치되서 수직 방향으로 출력된다.
pre_sale.groupby(pre_sale['연도']).describe().T
```

```python
# 데이터프레임의 특정 열을 지정할 때 
# 데이터프레임['열이름'] == 데이터프레임.열이름
pre_sale.groupby(pre_sale.연도).describe().T
```

전국 규모별 평균 분양 가격 

```python
# pivot_table(values, index, columns[,aggfunc, fill_value]) => aggfunc, fill_value는 생략이 가능하다.
# values : 통계 함수를 적용할 데이터프레임의 열을 지정한다.
# index : 그룹화의 첫 번째 기준으로 사용되는 피벗 테이블로 가져올 데이터프레임의 행을 지정한다. => 행 데이터
# columns : 그룹화의 두 번째 기준으로 사용되는 피벗 테이블로 가져올 데이터프레임의 열을 지정한다. => 열 데이터
# aggfunc : index와 columns로 그룹화된 values에 적용할 함수를 지정한다.
# => 기본값은 평균을 계산하는 함수 mean이다.
# fill_values : NaN을 대체할 값을 지정한다. => 주로 0을 사용한다.
pre_sale.pivot_table('평당분양가격', '규모구분', '연도', sum)
```

전국 지역별 평균 분양 가격

```python
# 규모구분이 전체로 되어있는 금액으로만 연도별 변동 금액을 살펴보자
# 규모구분이 전체인 데이터만 추출한다.
region_year_all = pre_sale[pre_sale['규모구분'] == '전체']
region_year_all
```

```python
# 피벗 테이블 결과는 2번째 인수(index)가 피벗 테이블의 index로 사용되므로 피벗 테이블의 인덱스를 다시 지정하려면
# reset_index() 함수를 사용해서 다시 index를 붙여주면 된다. 
# region_year = region_year_all.pivot_table('평당분양가격', '지역명', '연도')
region_year = region_year_all.pivot_table('평당분양가격', '지역명', '연도').reset_index()
region_year
```

전국 지역별 평균 분양가격 변동 금액

```python
region_year['변동액'] = region_year['2018'] - region_year['2015']
region_year
```

```python
max_delta_price = np.max(region_year['변동액'])
# region_year['변동액'].max()

print('2015년부터 2018년까지 분양가는 계속 상승했으며, 상승액이 가장 큰 지역은 제주로 상승액은 평당 {:,.1f}이다.'
      .format(max_delta_price))

# 문자열 출력시 줄바꿈을 할때에는 '...' |(역슬래시) '...' 사용
```

```python
# 최소
min_delta_price = np.min(region_year['변동액'])
min_delta_price
```

```python
# 평균
mean_delta_price = np.mean(region_year['변동액'])
mean_delta_price
```

```python
# 전국 지역별 평균 분양 가격 시각화
# ggplot(데이터프레임, aes(x='x축의 데이터', y='y축의 데이터', fill='범례'))
# geom_bar(state='identity') : 세로 막대 그래프, position 옵션을 지정하지 않으면 누적 막대그래프, 지정하면 막대 그래프가 작성된다.

ggplot(region_year_all, aes(x='지역명', y='평당분양가격', fill='연도')) \
    + geom_bar(stat='identity', position='dodge') \
    + ggtitle('2015 ~ 2018년 신규 민간 아파트 분양 가격') \
    + theme(text=element_text(family='NanumGothicCoding'), figure_size=(12, 8))
```
