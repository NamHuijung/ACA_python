# **1. Linear Layer**

* 신경망의 가장 기본 구성 요소
* Fully-connected(FC) Layer라고 불리기도 함
* 내부 파라미터에 따른 선형 변환을 수행하는 함수

# **2. Linear Layer 작동 방식**

* 각 입력 노드들에 weight(가중치)를 곱하고 모두 합친 뒤, bias(편향)을 더함
* 행렬 곱으로 구현 가능
* n차원에서 m차원으로 선형 변환 함수

<font color=red>* Linear Layer는 선형 변환 함수  
<font color=red>* 내부 가중치 파라미터(weight parameter) W와 b에 의해 정의  
<font color=red>* 이 함수의 파라미터를 잘 조절하면 주어진 입력에 대해 원하는 출력을 만들 수 있음</font>

```python
import torch
```

```python
W = torch.FloatTensor([[1, 2],
                       [3, 4],
                       [5, 6]])
b = torch.FloatTensor([2, 2])
```

```python
print(W.size())
print(b.size())
```

```python
def linear(x, W, b) :
  y = torch.matmul(x, W) + b
  return y
```

```python
x = torch.FloatTensor([[1, 1, 1],
                       [2, 2, 2],
                       [3, 3, 3],
                       [4, 4, 4]])
print(x.size())
```

```python
y = linear(x, W, b)
```

```python
print(y.size())
```

# **3. nn.Module**

```python
# nn(뉴럴 네트워크) 만들 때 상속받아서 사용하는 추상 클래스
import torch.nn as nn
```

```python
class MyLinear(nn.Module) :
  def __init__(self, input_dim=3, output_dim=2) :
    self.input_dim = input_dim
    self.output_dim = output_dim

    super().__init__()

    self.W = torch.FloatTensor(input_dim, output_dim)
    self.b = torch.FloatTensor(output_dim)

  def forward(self, x) :
    y = torch.matmul(x, self.W) + self.b
    return y
```

```python
linear = MyLinear(3, 2)
y = linear(x) # forward가 자동으로 호출.
```

```python
print(y.size())
```

```python
for p in linear.parameters() :
  print(p) # parameters 출력이 안됨.
```

### **nn.Parameter**

```python
class MyLinear(nn.Module) :
  def __init__(self, input_dim=3, output_dim=2) :
    self.input_dim = input_dim
    self.output_dim = output_dim

    super().__init__()

    self.W = nn.Parameter(torch.FloatTensor(input_dim, output_dim))
    self.b = nn.Parameter(torch.FloatTensor(output_dim))

  def forward(self, x) :
    y = torch.matmul(x, self.W) + self.b
    return y
```

```python
linear = MyLinear(3, 2)
y = linear(x)
```

```python
print(y.size())
```

```python
for p in linear.parameters() :
  print(p)
```

### **nn.Module의 Linear 함수를 통해 재구현**

```python
class MyLinear(nn.Module) :
  def __init__(self, input_dim=3, output_dim=2) :
    self.input_dim = input_dim
    self.output_dim = output_dim

    super().__init__()

    self.linear = nn.Linear(input_dim, output_dim)

  def forward(self, x) :
    y = self.linear(x)
    return y
```

```python
linear = MyLinear(3, 2)
y = linear(x)
```

```python
print(y.size())
```

```python
for p in linear.parameters() :
  print(p)
```
