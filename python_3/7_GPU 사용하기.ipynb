# **1. GPU 사용하는 방법**

```python
import torch
import torch.nn as nn
```

```python
x = torch.cuda.FloatTensor(2, 2) 
# 에러가 발생할 경우 런타임 -> 런타임 유형 변경에서 GPU를 선택
print(x)
# GPU가 여러개 있는 경우 GPU 번호를 확인
# 같은 디바이스에 있는 텐서끼리만 연산이 가능하다.
```

```python
x = torch.FloatTensor(2, 2)
print(x) # CPU(RAM) 사용 
```

```python
# GPU로 보냄
x.cuda()
```

```python
# GPU0으로 지정
d = torch.device('cuda:0')
```

```python
# move가 아니라 copy로 됨
x.cuda(device=d)
```

```python
x.device
# x가 CPU에도 있고 GPU에서 copy되어 있음.
```

# **2. Convert to CPU tensor from CUDA tensor**

```python
x = torch.cuda.FloatTensor(2, 2)
```

```python
x = x.cpu()
```

```python
d = torch.device('cpu')
x = x.to(d) # CPU를 CUDA로 바꿈
```

# **3. Move model from CPU to GPU**

```python
def print_params(model) :
  for p in model.parameters() :
    print(p)
```

```python
linear = nn.Linear(2, 2)
# 기본 Linear
print_params(linear)
```

```python
linear = linear.cuda()
# 쿠다에서 Linear
print_params(linear)
```

```python
# 다시 CPU로 옮기기
linear = linear.cpu()
print_params(linear)
```

```python
# GPU 번호까지 정해서 CUDA 바꾸기
d = torch.device('cuda:0')
linear = linear.to(d)
print_params(linear)
```
