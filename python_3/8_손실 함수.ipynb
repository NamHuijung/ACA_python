# **1. 손실 함수를 사용하는 목표**

* Linear Layer 함수가 얼마나 원하는 만큼 동작하는지 측정하고 싶음
* 잘 동작하는지 점수로 나타내고 싶음

### **Loss**

* 손실값 : 원하는 출력값(target, y)과 실제 출력값(output, ^y)의 차이의 합
* Loss가 작을수록 가상의 함수를 잘 모사하고 있다고 할 수 있음
* Loss가 작은 Linear Layer를 선택하면 됨

### **Loss Function**

* Linear Layer의 파라미터를 바꿀때마다 Loss를 계산

Conclusion
> 목표로 하는 함수를 모사하기 위해 학습용 입력 데이터들을 Linear Layer에 넣어 출력 값들을 구하고, 출력 값(^y)들과 목표 값(y)들의 차이의 합(Loss)을 최소화해야 함

### **Mean Square Error(MSE) Loss**

$$
\text{MSE}(\hat{x}_{1:N}, x_{1:N})=\frac{1}{N\times{n}}\sum_{i=1}^{N}{||x_i-\hat{x}_i||_2^2}
$$


```python
import torch
```

```python
def mse(x_hat, x) :
  y = ((x - x_hat) ** 2).mean()
  return y
```

```python
x = torch.FloatTensor([[1, 1],
                       [2, 2]])
x_hat = torch.FloatTensor([[0, 0],
                           [0, 0]])
print(x.size(), x_hat.size())
```

```python
mse(x_hat, x)
```

# **2. 파이토치로 손실 함수 사용**

```python
import torch.nn.functional as F
```

```python
F.mse_loss(x_hat, x)
```

```python
F.mse_loss(x_hat, x, reduction='sum')
```

```python
F.mse_loss(x_hat, x, reduction='none') # (x - x_hat) ** 2, 차원 축소X
```

```python
import torch.nn as nn
```

```python
mse_loss = nn.MSELoss()
mse_loss(x_hat, x)
```
